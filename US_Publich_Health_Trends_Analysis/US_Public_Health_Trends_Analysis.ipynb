# **Hypothesis Test**

## One sample Test

According to the EDA table 6, we want to test the below hypothesis at 0.05 level of significance:

*   Null Hypothesis (H0): The average proportion of High Blood Pressure(HighBP) cases in the Mississippi state is equal to or less than a specified national average .
*   Alternative Hypothesis (H1): The average proportion of High Blood Pressure(HighBP) cases in the Mississippi state is greater than the national average.


The null and alternate hypothesis are written as:<br>
<center> $H_0$ : $\mu \leq Average$ <br>
$H_1$ : $\mu$ $>$ $Average$

We used the $t$ distribution:<br>
<center> <center> $t$ = $\frac{\bar{X} - \mu}{s/\sqrt n}$</center>

<br>
where:<br>
$\bar{X}$ (sample mean) : is the average of the proportion values for a particular health issue across all states<br>
$\mu$ (hypothesized population mean):  is the value we are testing against,  a national average.<br>
$s$ (sample standard deviation): This measures the variability of the proportion values around the sample mean.<br>
$n$ (sample size): This is the number of observations in our sample, which is the count of states in dataset.<br>

**This is a one-tail test, and it is a upper tail test.**

because the alternate hypothesis (H1) specifies that the average proportion of High Blood Pressure(HighBP) cases in the Mississippi state is greater than the national average, which defines an upper direction.

**Decision Rule:**

Since this is a one-tail test, I will compare the calculated test statistic (t) to the critical value for the upper tail at the chosen significance level(0.05).
* If the absolute value of the test statistic (|t|) is greater than the critical value, you would reject the null hypothesis (H0).
* If the absolute value of the test statistic (|t|) is less than or equal to the critical value, you would fail to reject the null hypothesis (H0).

```r
# According to the table2, we calculated each points one by one.
# xbar calculation
library(dplyr)

numeric_data_Tb6 <- Table6 %>% select_if(is.numeric)

# xbar calculation
xbar <- mean(numeric_data_Tb6$HighBloodPressure,na.rm = TRUE)

# the hypothesized population mean
# REF https://www.cdc.gov/bloodpressure/facts.htm#:~:text=In%202021%2C%20hypertension%20was%20a,deaths%20in%20the%20United%20States.&text=Nearly%20half%20of%20adults%20have,are%20taking%20medication%20for%20hypertension
mu <- 48.1

# then calculate the sample std deviation
s <- sd(numeric_data_Tb6$HighBloodPressure,na.rm = TRUE)

# the size of sample
n <- length(numeric_data_Tb6$HighBloodPressure)

# we set the alpha equals to .05
alpha = 0.05

# Substitute in the formula:
t = (xbar - mu)/(s / sqrt(n))
round(t,3)

# Critical value calculation
t.alpha =qt(1-0.05, df=n-1)
t.alpha

# Calculate the p-value for a one-tailed test
pval = pt(t, df=n-1,lower.tail = F)
pval
```

**Decision:**<br>

Because the absolute critical value(1.663) is greater than the absolute t value, so we could conclude that H0 is NOT rejeccted.<br>
At the same time, the P value(1) is greater than the significance level value(0.05), so we also do NOT reject the H0 at the 0.05 significance level.<br>

```r
# then use plot to check our decision!
# rnorm definition
tmp <- rnorm(100000, mean=0, sd=1)

# density calculation
dens <- density(tmp)

# because the t value is far away the critical value,
# so we divided into two plots by two ranges
range1 <- c(-20, -12)
range2 <- c(-10, 10)

# setting layout
layout(matrix(1:2, nrow=1))

# plot for range1
plot(dens, main='Range -20 to -12', xlim=range1, xlab='', ylab='', type='n')
polygon(dens$x, dens$y, col='grey')
abline(v=t.alpha, col='red', lwd=3)
abline(v=t, lty=2, lwd=5, col='purple')

# plot for range2
plot(density(tmp), main='Upper Tail Test')
# add a line on the right tail
abline(v=t.alpha, col='red', lwd=3)
# add the shaded polygon to the RIGHT
dens = density(tmp) # store this to manipulate
x1 <- min(which(dens$x >= t.alpha))
x2 <- max(which(dens$x <= Inf))
with(dens, polygon(x=c(x[c(x1,x1:x2,x2)]), y= c(0, y[x1:x2], 0), col="skyblue"))

# now add our purple line
abline(v=t, lty=2, lwd=5, col='purple')

# reset layout
layout(1)
```

## Two Sample Test

We identified from our tables that south and midwest regions are the top two regions having greater average proportion of adults suffering from High BP. Among these two regions average proportions in south are greater than midwest. Hence, we decided to take two samples from these two regions and test whether the average proportions in south region is still greater than midwest region at 0.05 level of significance.<br>

In this case, we are interested in finding whether the average proportion in south is greater than the average proportion in midwest. <br>
Hence, this is **One Tail test (Upper-Tail)**

Null hypothesis($H_0$): The average proportion of HighBp cases in south are less than or equal to the average proportion in midwest
<br> <center>$H_0$ : $\mu_1 \leq \mu_2$</center><br>

Alternate hypothesis($H_1$):  The average proportion of HighBp cases in south are greater than the average proportion in midwest<br>

<center>$H_1$ : $\mu_1 > \mu_2$</center><br>

```r
# south region
set.seed(3127108)

# proportion of HighBP cases for south region
pop1= merged_df$HighBloodPressure[merged_df$Region=="South"]
s_pop1 = sd(pop1)
cat("standard deviation of pop1 is",s_pop1)

# Take a random sample from above data
sample1 = sample(pop1, size = 10, replace = FALSE)
Xbar1 = mean(sample1)
s1 = sd(sample1)
n1 = length(sample1)
```

```r
# midwest region
set.seed(3127109)
# proportion of HighBP cases for south region
pop2 = merged_df$HighBloodPressure[merged_df$Region=="Midwest"]
s_pop2 = sd(pop2)
cat("standard deviation of pop1 is",s_pop2)

# Take a random sample from above data
sample2 = sample(pop2, size = 10, replace = FALSE)
Xbar2 = mean(sample2)
s2 = sd(sample2)
n2 = length(sample2)
```

Here, popoulation standard deviations are not equal. But, we know two samples standard deviations and variances and they are not equal. therefore, we consider t test statistic  <br>
The formula for the test statistic $t$ for unequal variances:<br>
<center> $t$ = $\frac{{\bar{X_1}-\bar{X_2}}}{\sqrt{\frac{s^2_1}{n_1} + \frac{s^2_2}{n_2}}}$ <br></center><br>

where:<br>
$\bar{X_1}$ is the mean of the first sample<br>
$\bar{X_2}$ is the mean of the second sample<br>
$n_1$ is the number of observations in the first sample<br>
$n_2$ is the number of observations in the second sample<br>
$s_1^2$ is the variance of the first sample<br>
$s_2^2$ is the variance of the second sample<br>

The degrees of freedom for unequal variance test is found by:<br>
<center> $df$ = $\frac{[(\frac{s_1^2}{n_1}) + (\frac{s_2^2}{n_2})]^2}{\frac{(\frac{s_1^2}{n_1})^2}{n_1 - 1} + \frac{(\frac{s_2^2}{n_2})^2}{n_2 - 1}}$</center><br>

```r
# degrees of freedom
df = (((s1^2/n1) + (s2^2/n2))^2)/((((s1^2/n1)^2)/(n1-1)) + ((s2^2/n2)^2/(n2-1)))
df
```

```r
alpha=0.05 # level of significance

crl_val = qt(1-alpha, df=df)
cat("critical value:",crl_val)
```

**Decision rule:**<br>
 If $t$ value is less than $1.77$ $\rightarrow$ do not reject the null hypothesis($H_0$).<br>
 If $t$ value is greater than $1.77$ $\rightarrow$ reject the null hypothesis($H_0$).

```r
# t test statistic
t = round((Xbar1 - Xbar2)/sqrt((s1^2/n1) + (s2^2/n2)),2)
cat("t value:",t,"\n")

# p-value
pval = pt(t,df)
cat("p-value:",pval)
```

**Decision:** Computed $t$ value $4.45$ is greater than $1.77$ and falls in the area of rejection; that is, it is to the right of $1.77$. Therefore, the null hypothesis($H_0$) is rejected at the $0.05$ significance level. <br>




```r
# plot
tmp = rnorm(10000000, mean=0, sd=1)
plot(density(tmp), main='Upper Tail Test')

# add a red line on the right tail for critical value
abline(v=crl_val, col='red', lwd=3)
# add the shaded polygon to the RIGHT
dens = density(tmp) # store this to manipulate
x1 <- min(which(dens$x >= crl_val))
x2 <- max(which(dens$x <= Inf))
with(dens, polygon(x=c(x[c(x1,x1:x2,x2)]), y= c(0, y[x1:x2], 0), col="skyblue"))

# add a purple line for t value
abline(v=t, lty=2, lwd=5, col='purple')

# legends
legend("topleft", legend = c("critical value", "z value"), col = c("red", "purple"), lty = c(1, 2))
```

**Interpretation:**The average proportion of HighBP cases in south region is greater than the average proportion in midwest region
